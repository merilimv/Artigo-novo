```python
import pandas as pd
import numpy as np
```

```python
adult = pd.read_csv("preprocessed_adult.csv")

print(adult.shape)

adult_copy = adult.drop(['education_10th', 'education_11th', 'education_12th',
       'education_1st-4th', 'education_5th-6th', 'education_7th-8th',
       'education_9th', 'education_Assoc-acdm', 'education_Assoc-voc',
       'education_Bachelors', 'education_Doctorate', 'education_HS-grad',
       'education_Masters', 'education_Preschool', 'education_Prof-school',
       'education_Some-college'], axis = 1)
# education se refere ao nível de escolaridade da pessoa, que é uma categoria ordinal
# tem outra featura chamada education_num que é um ordinal encoding dessa, então não tem necessidade de manter essas

classes_to_int = adult_copy.columns[8:]

adult_copy[classes_to_int] = adult_copy[classes_to_int].astype(int)

adult_copy.info()
```

```python
adult_copy['marital-status_Married-civ-or-AF-spouse'] = adult_copy['marital-status_Married-civ-spouse'] | adult_copy['marital-status_Married-AF-spouse']
adult_copy = adult_copy.drop(['marital-status_Married-AF-spouse', 'marital-status_Married-civ-spouse'], axis = 1)
# só existe cerca de 30 pessoas com a classe Married_AF_spouse, e elas são semelhantes em dist. ao Married-civ-spouse
# então resolvi juntar para reduzir a dimensionalidade, outra opção seria só retirar essas 30
```

```python
countrys = adult_copy.columns[45:86] # pegando as colunas dos native-country
```

```python
adult_copy['region_US-and-Cuba'] = adult_copy['native-country_United-States'] | adult_copy['native-country_Cuba']

adult_copy['region_Western'] = adult_copy['native-country_England'] | adult_copy['native-country_Germany'] | \
                               adult_copy['native-country_Canada'] | adult_copy['native-country_Italy'] | \
                               adult_copy['native-country_France'] | adult_copy['native-country_Greece'] | \
                               adult_copy['native-country_Philippines']

adult_copy['region_ThirdWorld-and-Portugal'] = adult_copy['native-country_Mexico'] | adult_copy['native-country_Puerto-Rico'] | \
                                               adult_copy['native-country_Honduras'] | adult_copy['native-country_Jamaica'] | \
                                               adult_copy['native-country_Columbia'] | adult_copy['native-country_Laos'] | \
                                               adult_copy['native-country_Portugal'] | adult_copy['native-country_Haiti'] | \
                                               adult_copy['native-country_Dominican-Republic'] | adult_copy['native-country_El-Salvador'] | \
                                               adult_copy['native-country_Guatemala'] | adult_copy['native-country_Peru'] | \
                                               adult_copy['native-country_Trinadad&Tobago'] | adult_copy['native-country_Outlying-US(Guam-USVI-etc)'] | \
                                               adult_copy['native-country_Nicaragua'] | adult_copy['native-country_Vietnam'] | \
                                               adult_copy['native-country_Holand-Netherlands']

adult_copy['region_Eastern'] = adult_copy['native-country_India'] | adult_copy['native-country_Iran'] | \
                               adult_copy['native-country_Cambodia'] | adult_copy['native-country_Taiwan'] | \
                               adult_copy['native-country_Japan'] | adult_copy['native-country_Yugoslavia'] | \
                               adult_copy['native-country_China'] | adult_copy['native-country_Hong']

adult_copy['region_Others'] = adult_copy['native-country_South'] | adult_copy['native-country_Poland'] | \
                              adult_copy['native-country_Ireland'] | adult_copy['native-country_Hungary'] | \
                              adult_copy['native-country_Scotland'] | adult_copy['native-country_Thailand'] | \
                              adult_copy['native-country_Ecuador']

# aqui eu juntei os países por região, mas também levei em consideração a distribuição deles no dataset, tentando manter dist parecidas juntas

adult_copy = adult_copy.drop(countrys, axis = 1)
```

```python
adult_copy = adult_copy.drop('fnlwgt', axis = 1)
# dropando o fnlwgt que é geralmente não-utilizado

adult_copy.info() 
# resultou em 50 colunas, tinhamos 105 antes, poderia tirar mais mas acho que não há tanta necessidade
```

```python
adult_copy.to_csv("binned_adult.csv", index = False)
```
